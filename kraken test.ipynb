```
from kraken import binarization, pageseg, rpred
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# ===== 1. Đọc ảnh gốc =====
im = Image.open('/content/sample_img.jpg')

# ===== 2. Binarize (chuyển ảnh sang trắng đen) =====
bin_im = binarization.nlbin(im)
bin_im.save('binarized.png')

# ===== 3. Segment (phân vùng văn bản) =====
seg_model_path = "/root/.local/share/htrmopo/97665cf3-f83d-5594-8855-f28d3af9df7a/blla.mlmodel"
segments = pageseg.segment(bin_im, model=seg_model_path)

# ===== 4. Vẽ vùng segmentation lên ảnh =====
draw = ImageDraw.Draw(bin_im)
for line in segments['lines']:
    draw.polygon(line['baseline'], outline='red')

# Lưu và hiển thị ảnh segmentation
bin_im.save('segmented.png')
plt.imshow(bin_im)
plt.axis('off')
plt.show()

# ===== 5. OCR with McCATMuS model =====
ocr_model_path = "/root/.local/share/htrmopo/78d7d26e-50b3-55f8-b839-c7a882ea2053/McCATMuS_nfd_nofix_V1.mlmodel"
ocr = rpred.TorchSeqRecognizer.load_from_filename(ocr_model_path)

pred = rpred.rpred(ocr, bin_im, segments)
text = '\n'.join([line.text for line in pred])

# ===== 6. Export result =====
print("📜 OCR Result:\n")
print(text)

with open('result.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```
